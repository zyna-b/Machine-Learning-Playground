```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
# Lets apply OLS (Odinary Least Square to get to know about the values about our m and b which are our paramerters i.e., our tuning nobes)
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

# Slope: m

```python
reg.coef_
```

# Intercept: b

```python
reg.intercept_
```

Now we want to predict the best fit line using gradient descent and try to get close to these m and b values.

# Gradient Descent

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
```

```python
# Lets apply gradient descent asumming slope: m is constant m = 78.35
# and let's assume the starting value of intercept i.e., b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#00a2e8', label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Lets take the learning rate = 0.1
lr = 0.1

step_size = lr * loss_slope
step_size
```

```python
# Calculating the new intercept i.e., b
b = b - step_size
b
```

```python
# Again calculate the slope of new b and decide and new b

y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred1, color='#00a2e8', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#00a998', label='b = 0')
plt.legend()
plt.show
```

```python
# 3rd iteration now
loss_slope = -2 * np.sum(y - m*X.ravel() - b)   # Here the value of b is the new b that we calculated
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
```

```python
# Again calculate the slope of new b and decide and new b

y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred1, color='#00a2e8', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#00a998', label='b = 0')
plt.plot(X, y_pred2, color='#009999', label='b = {}'.format(b))
plt.legend()
plt.show
```

```python
# Iteration 4
loss_slope = -2 * np.sum(y - m*X.ravel() - b)   # Here the value of b is the new b that we calculated
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
# Again calculate the slope of new b and decide and new b

y_pred3 = ((78.35 * X) + b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred1, color='#00a2e8', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#00a998', label='b = 0')
plt.plot(X, y_pred2, color='#005549', label='b = {}'.format(b))
plt.plot(X, y_pred3, color='#009000', label='b = {}'.format(b))
plt.legend()
plt.show
```

# We can put all this in a loop

```python
b = -100
m = 78.35
lr = 0.01
epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() -b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X, y_pred)
plt.scatter(X, y)
```
